# Per Capita Income of Each State of India Via Web Scrapping

Import Libraries


```python
import os
import requests
from bs4 import BeautifulSoup
```

Taken the web URL from where the Data is gonna be fetched

```python
url = 'https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP_per_capita'
response = requests.get(url)
content = response.content
```

Parsed the whole content from the URL into BeautifulSoup in order to scrap the Per Capita Income 

```python
soup = BeautifulSoup(content, 'html.parser')
table = soup.find('table', class_='wikitable')


data = []
for row in table.tbody.find_all('tr'):
    columns = row.find_all('td')
    if len(columns) >= 3:
        state = columns[1].text.strip()
        income = columns[2].text.strip().replace('â‚¹', '').replace(',', '')
        data.append((state, int(income)))
```

Displaying Data

```python
data
```

Taken the whole fetched data into Pandas Data Frame

```python
import pandas as pd
DataTable = pd.DataFrame(data, columns = ['State', 'PCI'])
```

A .shp image file of Indian Map in which the Per Capita will be shown.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

shapefile_path = 'D:\\Newbieron Internship Tasks\\Task 3\\states_india\\states_india.shp'
map_df = gpd.read_file(shapefile_path)

DataTable['State'] = DataTable['State'].replace('Jammu and Kashmir', 'Jammu & Kashmir')
```

```python
# Merge the shapefile with the per capita income data
merged = map_df.set_index('st_nm').join(DataTable.set_index('State'))
```

```python
# Set up the figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Create the choropleth map with a custom color scheme
merged.plot(column='PCI', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add state names as labels
merged['centroid'] = merged.geometry.centroid
for x, y, label in zip(merged['centroid'].x, merged['centroid'].y, merged.index):
    ax.text(x, y, label, fontsize=8, ha='center')


# Set the chart title
plt.title('Per Capita Income of Indian States')

# Display the chart
plt.show()
```
